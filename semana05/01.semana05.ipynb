## Valores aleatórios para testarmos

```python
# Criando valores aleatórios de média 5 e desvio padrão 3
x_10 = sct.norm.rvs(loc=5, scale=3, size=10)
x_50 = sct.norm.rvs(loc=5, scale=3, size=50)
x_100 = sct.norm.rvs(loc=5, scale=3, size=100)
x_1000 = sct.norm.rvs(loc=5, scale=3, size=1000)
x_10000 = sct.norm.rvs(loc=5, scale=3, size=10000)
x_100000 = sct.norm.rvs(loc=5, scale=3, size=100000)
```

## Histograma

```python
plt.hist(x_10)
```

## Shapiro

```python
print(sct.shapiro(x_10))
print(sct.shapiro(x_50))
print(sct.shapiro(x_100))
print(sct.shapiro(x_1000))
print(sct.shapiro(x_10000))
print(sct.shapiro(x_100000))
```

## QQPlot

```python
import statsmodels.api as sm
sm.qqplot(x_10,line='s')
sm.qqplot(x_50, line='s')
sm.qqplot(x_100, line='s')
sm.qqplot(x_1000, line='s')
plt.show()
```

## Jarque

```python
print(sct.jarque_bera(x_10))
print(sct.jarque_bera(x_50))
print(sct.jarque_bera(x_100))
print(sct.jarque_bera(x_1000))
print(sct.jarque_bera(x_10000))
print(sct.jarque_bera(x_100000))
```

## Normal Test

```python
print(sct.normaltest(x_10))
print(sct.normaltest(x_50))
print(sct.normaltest(x_100))
print(sct.normaltest(x_1000))
print(sct.normaltest(x_10000))
print(sct.normaltest(x_100000))
```
