# Regressão Linear

```python
import seaborn as sns
```

```python
import pandas as pd
import numpy as np

df_train = pd.read_csv('train.csv').fillna(0)
df_test = pd.read_csv('test.csv').fillna(0)

x_train = df_train['x']
y_train = df_train['y']
x_test = df_test['x']
y_test = df_test['y']

x_train = np.array(x_train)
y_train = np.array(y_train)
x_test = np.array(x_test)
y_test = np.array(y_test)

x_train = x_train.reshape(-1,1)
x_test = x_test.reshape(-1,1)
```

```python
df_train
```

```python
sns.scatterplot(x='x',y='y', data=df_train[df_train.x < 3500])
```

```python
sns.scatterplot(x='x',y='y', data=df_test)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.linear_model import LinearRegression ,Ridge
from sklearn.metrics import r2_score
clf = LinearRegression(normalize=True)
clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
print(r2_score(y_test,y_pred))
```

```python
clf = Ridge(normalize=True)
clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
print(r2_score(y_test,y_pred))
```

```python
df_pred = df_test
df_pred['new_y'] = y_pred
df_pred
```

```python
sns.scatterplot(x='x',y='new_y', data=df_pred,color='red')
sns.scatterplot(x='x',y='y', data=df_pred,color='blue')
```
